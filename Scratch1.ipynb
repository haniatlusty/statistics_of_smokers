```python
sns.histplot(df["age"], stat = "density", color = "lightgreen", kde = True)
plt.vlines(max(df["age"]), ymin = 0, ymax = 0.02, color = "red", label = "Max age: {}".format(max(df["age"])))
plt.vlines(min(df["age"]), ymin = 0, ymax = 0.02, color = "yellow", label = "Min age: {}".format(min(df["age"])))
plt.title("The histogram of smoker's age")
plt.vlines(np.mean(df["age"]), ymin = 0, ymax = 0.02, label = "Average age: {}".format(int(np.mean(df["age"]))))

plt.legend()
```

```python
x = np.linspace(-1,1)
sns.boxplot(y = df["age"], color = 'lightgreen')
plt.plot(x, [max(df["age"])]*50, color = "red", label = "max")
plt.plot(x, [min(df["age"])]*50, color = "orange", label = "min")
plt.plot(x, [np.median(df["age"])]*50, color = "green", label = "median")
plt.plot(x, [np.quantile(df["age"], 0.25)]*50, color = "blue", label = "quantile 0.25")
plt.plot(x, [np.quantile(df["age"], 0.75)]*50, color = "pink", label = "quantile 0.75")
plt.title("Boxplot that includes the description of basic statistics")
plt.legend()
```
