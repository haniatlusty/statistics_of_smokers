```python
sns.histplot(df["age"], stat = "density", color = "lightgreen", kde = True)
plt.vlines(max(df["age"]), ymin = 0, ymax = 0.02, color = "red", label = "Max age: {}".format(max(df["age"])))
plt.vlines(min(df["age"]), ymin = 0, ymax = 0.02, color = "yellow", label = "Min age: {}".format(min(df["age"])))
plt.title("The histogram of people's age")
plt.vlines(np.mean(df["age"]), ymin = 0, ymax = 0.02, label = "Average age: {}".format(int(np.mean(df["age"]))))

plt.legend()
```

```python
x = np.linspace(-1,1)
sns.boxplot(y = df["age"], color = 'lightgreen')
plt.plot(x, [max(df["age"])]*50, color = "red", label = "max")
plt.plot(x, [min(df["age"])]*50, color = "orange", label = "min")
plt.plot(x, [np.median(df["age"])]*50, color = "green", label = "median")
plt.plot(x, [np.quantile(df["age"], 0.25)]*50, color = "blue", label = "quantile 0.25")
plt.plot(x, [np.quantile(df["age"], 0.75)]*50, color = "pink", label = "quantile 0.75")
plt.title("Boxplot that includes the description of basic statistics")
plt.legend()
```

```python
unique_columns = ["gender", "marital_status", "highest_qualification", "nationality", "ethnicity", "region"]
```

```python
uniques = pd.DataFrame()
for column in unique_columns:
    uni = df[column].unique()
    df_uni = pd.DataFrame(uni, columns = [column])
    uniques = pd.concat([uniques, df_uni], axis = 1)
uniques
```

And now let's check if the unique values are similar to smokers.

```python
uniques_smokers = pd.DataFrame()
for column in unique_columns:
    uni = smokers[column].unique()
    df_uni = pd.DataFrame(uni, columns = [column])
    uniques_smokers = pd.concat([uniques_smokers, df_uni], axis = 1)
uniques_smokers
```

```python
uniques["gender"].isin(uniques_smokers["gender"])
```

```python
uniques["highest_qualification"].isin(uniques_smokers["highest_qualification"]).values
```

```python
i = 0
for col in unique_columns:
    if True in uniques[col].isin(uniques_smokers[col]).values :
        print("{} column are both the same".format(col))
        i = i + 1
print("{} of {} colums comparing are the same".format(i, len(unique_columns)))
```
